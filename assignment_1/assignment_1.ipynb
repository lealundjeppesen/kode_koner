````python
import pandas as pd 

#Defining the function
def process_weather(df):
    
    #We do not need iloc[:,:4] because there is only 4 cols in the csv file. Also, we do not need to load the data as this is in a local csv file that will be converted to a datafram and inserted.

    # structure and clean data using methods chaining. Sorting by station asc and datetime desc.
    df_out = \
        df\
            .rename(columns={0: 'station', 1: 'datetime', 2: 'obs_type', 3: 'obs_value'})\
            .query("obs_type == 'TMAX'")\
            .assign(obs_value=lambda df: df['obs_value']/10)\
            .sort_values(by=['station', 'datetime'], ascending=[True, False])\
            .reset_index(drop=True)\
            .copy() 

    # area process
    df_out['area'] = df_out['station'].str[0:2]

    # datetime process
    df_out['datetime_dt'] = pd.to_datetime(df_out['datetime'], format = '%Y%m%d')
    df_out['month'] = df_out['datetime_dt'].dt.month
    df_out['year'] = df_out['datetime_dt'].dt.year

    return df_out

#Putting in the file path
file_path='./data/weather_data_1870-1875.csv'
#Reading the csv file as a pandas
df=pd.read_csv(file_path, header=None)

#Use the defined function to structure and clean the dataframe
df_weather_period=process_weather(df)

#Print the dataframe
df_weather_period
````

````python
def get_colspecs(lines):
    """Extracts colspecs from `ghcnd-stations-column-metadata.txt`.
    
    Args:
        lines (list[str]): 
            list of relevant rows from `ghcnd-stations-column-metadata.txt` 
    
    Returns:
        (list[str]): 
            list of extracted colspecs i.e. ['1-11', '13-20', ..., '81-85']
    """
    colspec_idx_start = 13
    colspec_idx_end = 17 + 1  # Including idx 17
    # Extract colspecs from each line
    colspecs = [line[colspec_idx_start:colspec_idx_end].strip() for line in lines]

    return colspecs


# Apply get_colspecs to lines
colspecs = get_colspecs(lines)

# Print colspecs to check
for colspec in colspecs:
    print(colspec)
````

> **Ex. 3.3.3.4:** Write a function named `get_colspec_pair` which takes as input a string variable named `colspec` and returns a tuple of integers. Specifically, the function should take a string similar to each element of `colspecs`, split this string by `-` and return a tuple of integers where
1. The first integer should have `1` subtracted from it (Python is 0-indexed!)
2. The second integer should stay as it is (the intervals provided to the pandas function `read_fwf` should be non-inclusive)
> As an example, applying the function to `"1-11"` and `"13-20"` should yield the following results:

```python
print(get_colspec_pair("1-11"))
## output: (0, 11)

print(get_colspec_pair("13-20"))
## output: (12, 20)
```


````python
def get_colspec_pair(colspec):
    # Split the colspec by '-'
    parts = colspec.split('-')

    # Convert both parts to int, subtract 1 from the first part
    start = int(parts[0]) - 1
    end = int(parts[1])

    # Return as tuple
    return (start, end)

# Test the function
print(get_colspec_pair("1-11"))  # output: (0, 11)
print(get_colspec_pair("13-20"))  # output: (12, 20)
````

> **Ex. 3.3.3.6:** Because the fixed width column information spans the interval from `13` up and including index `17`, we know that the entries from `0` to `13` (non-inclusive) are the column names and the entries from `18` to the end of each line are the data types. Likewise, apply the function `get_column_dtypes` to the `lines` variable and store the output in a variable named `column_dtypes`.

````python
def get_column_names(lines):
    """Extracts column names from lines.
    
    Args:
        lines (list[str]): Lines to extract column names from.
    
    Returns:
        list[str]: Extracted column names.
    """
    # Extract column names from each line
    column_names = [line[:13].strip() for line in lines]

    return column_names

def get_column_dtypes(lines):
    """Extracts column data types from lines.
    
    Args:
        lines (list[str]): Lines to extract column data types from.
    
    Returns:
        list[str]: Extracted column data types.
    """
    # Extract column data types from each line
    column_dtypes = [line[18:].strip() for line in lines]

    return column_dtypes

# Apply get_column_names to lines
column_names = get_column_names(lines)

# Apply get_column_dtypes to lines
column_dtypes = get_column_dtypes(lines)

# Print column_names and column_dtypes to inspect
print("Column Names:", column_names)
print("Column Data Types:", column_dtypes)
````
